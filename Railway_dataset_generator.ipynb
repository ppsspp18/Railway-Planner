```python
import json
import random
from datetime import datetime, timedelta

START_DATE = datetime(2026, 4, 1)
DAYS_TO_GENERATE = 10
TOTAL_SEATS_PER_TRAIN = 350

def generate_seat_inventory():
    try:
        with open("train_dataset.json", "r") as f:
            trains = json.load(f)
    except FileNotFoundError:
        print("Error: Make sure 'train_dataset.json' is in the same folder!")
        return

    seat_dataset = []

    for day_offset in range(DAYS_TO_GENERATE):
        current_date = START_DATE + timedelta(days=day_offset)
        date_str = current_date.strftime("%Y-%m-%d")

        for train in trains:
            leg1 = f"{train['src']}-{train['first_stop']}"
            leg2 = f"{train['first_stop']}-{train['second_stop']}"
            leg3 = f"{train['second_stop']}-{train['dest']}"

            is_weekend = current_date.weekday() >= 5

            if is_weekend:
                min_seats, max_seats = -100, 150
            else:
                min_seats, max_seats = -50, 350

            inventory_record = {
                "train_number": train["train_number"],
                "journey_start_date": date_str,
                "leg_availability": {
                    leg1: random.randint(min_seats, max_seats),
                    leg2: random.randint(min_seats, max_seats),
                    leg3: random.randint(min_seats, max_seats)
                }
            }
            seat_dataset.append(inventory_record)

    with open("seat_dataset.json", "w") as f:
        json.dump(seat_dataset, f, indent=4)

    print(f"Successfully generated seat inventory for {len(trains)} trains over {DAYS_TO_GENERATE} days!")
    print(f"Total inventory records generated: {len(seat_dataset)}")

if __name__ == "__main__":
    generate_seat_inventory()
```
